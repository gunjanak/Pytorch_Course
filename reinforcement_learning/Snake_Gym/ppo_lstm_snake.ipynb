```python
import gym
import gym
import numpy as np
import torch

from snake999 import SnakeEnvWithPenalty

from ppo_lstm import PPOAgent

import pygame
# Initialize Pygame
pygame.init()
pygame.font.init()
```

```python
HEIGHT = 10
WIDTH = 10
env = SnakeEnvWithPenalty(width=600, height=600, rows=HEIGHT, cols=HEIGHT)
agent = PPOAgent(height=HEIGHT,width=WIDTH)
checkpoint = 'ppo_lstm_snake999_checkpoint.pth'
```

```python
try:
    agent.load(checkpoint)
except Exception as e:
    print(e)
```

```python
agent.train(env,num_episodes=5000, checkpoint_path=checkpoint)
```

```python
agent.plot("ppo_lstm_snake999.png")
```

```python
agent.test(env)
```
